```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%d-%m-%Y')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df['day_name'] = df['date'].dt.day_name()
df
```

```python
map = {
    'sun': 1,
    'fog': 2,
    'drizzle': 3,
    'rain': 4,
    'snow': 5
}
#berdasarkan tingkat kecerahan
```

```python
df['weather_numerik'] = df['weather'].map(map)
df
```

```python
Df = df.drop(columns=['date','weather','year','month','day','day_name'])
```

```python
Df
```

```python
from imblearn import under_sampling, over_sampling

X = Df.drop(columns=['weather_numerik'])
y = Df['weather_numerik']

# Undersampling
X_under, y_under = under_sampling.RandomUnderSampler().fit_resample(X, y)

# Overampling
X_over, y_over = over_sampling.RandomOverSampler().fit_resample(X, y)

# Overampling with smote
X_over_smote, y_over_smote = over_sampling.SMOTE().fit_resample(X, y)
```

```python
df_undersampling = pd.concat([X_under, y_under], axis=1)
df_oversampling = pd.concat([X_over, y_over], axis=1)
df_oversampling_smote = pd.concat([X_over_smote, y_over_smote], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_test)
X_test_std = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=32)
logreg.fit(X_train_std, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=32)
dt.fit(X_train_std,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=32)
model.fit(X_train_std, y_train)
preds = model.predict(X_test_std)
```
